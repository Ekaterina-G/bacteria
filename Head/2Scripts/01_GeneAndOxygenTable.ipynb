```python
import pandas as pd
from Bio import SeqIO
import os
```

```python
direct = '/home/emulciber/MitoClub/bacteria/Body/1Raw/corynebacteria_genes_aa'
dirname = os.listdir(direct)
```

```python
bacteria_test = pd.DataFrame()
```

```python
species = []
gene = []
aaseq = []


for file in dirname:
    try:
        for seq_record in SeqIO.parse((os.path.join(direct, file)), 'fasta'):
            gene.append(file.split('.')[0]) # сюда воткнуть название файла
            aaseq.append(str(seq_record.seq)) # получение последовательности из фасты
            # получение вида из дескрипшна
            desc = seq_record.description
            if str(desc.split('[')[-1]).startswith('Ga') == True:
                species.append(' '.join(desc.split('[')[-1].split(' : ')[-1].split(' ')[0:2]))
            else:
                species.append(' '.join(desc.split('[')[-1].split(' ')[0:2]))
    except:
        pass
    
    
bacteria_test['Species'] = species
bacteria_test['Gene'] = gene
bacteria_test['Sequence'] = aaseq
bacteria_test
```

```python
bacteria_test.to_csv('/home/emulciber/MitoClub/bacteria/Body/2Derived/bacteria_test.csv', sep='\t')
```
